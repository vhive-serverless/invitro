```python
import pandas as pd
```

# load preprocessed_150

```python
base_invocation_trace_name = "data/traces/reference/preprocessed_150/invocations.csv"
# base_invocation_trace_name = "data/traces/reference/sampled_150/4000/invocations.csv"
base_invocation_trace = pd.read_csv(base_invocation_trace_name)
display(base_invocation_trace.head())
```

```python
# get total invocation count for each timestamp
# drop HashOwner, HashApp, HashFunction, and Trigger column

invocation_count_only_stats = pd.DataFrame()
invocation_count_only = base_invocation_trace.drop(columns=["HashOwner", "HashApp", "HashFunction", "Trigger"])
invocation_count_only_stats["invocation_count_sum"] = invocation_count_only.sum(axis=1)
invocation_count_only_stats["invocation_count_avg"] = invocation_count_only.mean(axis=1)
invocation_count_only_stats["invocation_count_max"] = invocation_count_only.max(axis=1)
invocation_count_only_stats["invocation_count_min"] = invocation_count_only.min(axis=1)
# get normalized


display(invocation_count_only_stats.head())

invocation = pd.DataFrame()
invocation = pd.concat([base_invocation_trace[["HashOwner", "HashApp", "HashFunction", "Trigger"]], invocation], axis=1)
invocation = pd.concat([invocation, invocation_count_only_stats], axis=1)
invocation = pd.concat([invocation, invocation_count_only], axis=1)
invocation = invocation.sort_values(by=["invocation_count_sum"], ascending=False)

# filter number of invocations min and max
# If target RPS is 50, then min should be 600 (50*60/5) and max should be 6000 (50*60*2)
invocation_filtered = invocation[(invocation["invocation_count_max"] < 6000) & (invocation["invocation_count_min"] > 600)]

display(len(invocation_filtered))
display(invocation_filtered.head(30))
```

```python
def get_function_per_target_rps(invocation: pd.DataFrame, target_rps: int|float) -> pd.DataFrame:
    # target_rps = 50
    target_invocation_min = target_rps * 60 / 10
    target_invocation_max = target_rps * 60 * 2
    invocation_filtered = invocation[(invocation["invocation_count_max"] < target_invocation_max) & (invocation["invocation_count_min"] > target_invocation_min)].copy()
    if len(invocation_filtered) == 0:
        raise ValueError(f"No function found for target_rps {target_rps}")
    
    # get invocation that invocation_count_avg is closest to target_rpm
    target_rpm = target_rps * 60
    invocation_filtered["invocation_count_avg_diff"] = abs(invocation_filtered["invocation_count_avg"] - target_rpm)
    invocation_filtered = invocation_filtered.sort_values(by=["invocation_count_avg_diff"])
    return invocation_filtered.iloc[0:1]

filtered_trace = get_function_per_target_rps(invocation, 50)
display(filtered_trace)
```

```python
# this rps is the load where each workload utilizes a single node to 50%.
# so, if we deploy 1 function per workload, we will need at least 5 nodes to run them.

# since we want to increment the number of function deployed to measure deployment density, divide this rps by 10.
# then, deploying 1 function per workload will utilize 5% of a node.
# so, then, we will have utilizing 0.5 node, and deploying 2 functions per workload will utilize 1 node.
# with 4 nodes (current setup), we can deploy 8 functions per workload at max. Expecting around 5 functions.
per_workload_target_rps = {
    "chameleonserve" : 1000,
    "cnnserve" : 120,
    "imageresize": 24,
    "lrserving": 950,
    "mapper": 75,
    "pyaesserve": 1400,
    "reducer": 16,
    "rnnserve": 600,
    "streducer": 320,
    "sttrainer": 275,
}
```

```python
per_workload_trace = pd.DataFrame()
for workload, target_rps in per_workload_target_rps.items():
    filtered_trace = get_function_per_target_rps(invocation, target_rps / 10)
    # Add workload and target_rps as columns to the filtered trace
    filtered_trace_with_info = filtered_trace.copy()
    filtered_trace_with_info.insert(0, 'FunctionName', workload)
    filtered_trace_with_info.insert(1, 'target_rps', target_rps * 6)
    
    per_workload_trace = pd.concat([per_workload_trace, filtered_trace_with_info], ignore_index=True)

display(per_workload_trace)
```

```python
def generate_trace(per_workload_trace: pd.DataFrame, function_multiplier: int) -> pd.DataFrame:
    generated_trace = pd.DataFrame()
    
    # Get numbered columns (minute timestamps)
    numbered_columns = [col for col in per_workload_trace.columns if col.isdigit()]
    trace_length = len(numbered_columns)
    
    
    for index, row in per_workload_trace.iterrows():
        for i in range(function_multiplier):
            new_row = row.copy()
            new_row["FunctionName"] = f"{row['FunctionName']}"
            
            # Apply time shift with wrap-around
            if i > 0:
                shift_amount = 1 * i % trace_length
                # Get the invocation values for numbered columns
                invocation_values = [new_row[col] for col in numbered_columns]
                # Shift the values (rotate left by shift_amount)
                shifted_values = invocation_values[shift_amount:] + invocation_values[:shift_amount]
                # Update the row with shifted values
                for j, col in enumerate(numbered_columns):
                    new_row[col] = shifted_values[j]
            
            generated_trace = pd.concat([generated_trace, pd.DataFrame([new_row])], ignore_index=True)
    
    # Keep only FunctionName and numbered columns
    generated_trace = generated_trace[["FunctionName"] + numbered_columns]
    return generated_trace

generated_trace = generate_trace(per_workload_trace, 2)
display(generated_trace)
generated_trace.to_csv("data/traces/nexus/invocations.csv", index=False)
```
